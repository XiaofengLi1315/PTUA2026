```python
# magic command to display matplotlib plots inline within the ipython notebook webpage
%matplotlib inline

# import necessary modules
import pandas as pd, numpy as np, matplotlib.pyplot as plt
import matplotlib.cm as cm, matplotlib.font_manager as fm
```

```python
# load the gps coordinate data, using the date as the full set's index
# the data files are encoded as utf-8: specify so to prevent matplotlib from choking on diacritics
df = pd.read_csv('C:/Users/lxf/Desktop/PTUA-lab2/data lab5/summer-travel-gps-full.csv', encoding='utf-8', index_col='date', parse_dates=True)
rs = pd.read_csv('C:/Users/lxf/Desktop/PTUA-lab2/data lab5/summer-travel-gps-dbscan.csv', encoding='utf-8')
```

```python
# 统计每个国家被访问的次数，取前 10 名
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10, 6))

plt.bar(
    top_countries.index,
    top_countries.values,
    width=0.6,
    alpha=0.8,
    color='steelblue',
    edgecolor='black'
)

plt.title('Most Visited Countries (Summer Travel)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Number of Visits')

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.ylim(0, top_countries.values.max() * 1.1)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 统计最常访问的前 10 个城市
top_cities = df['city'].value_counts().head(10)

plt.figure(figsize=(10, 6))

plt.bar(
    top_cities.index,
    top_cities.values,
    width=0.6,
    alpha=0.8,
    color='coral',
    edgecolor='black'
)

plt.title('Most Visited Cities (Summer Travel)', fontsize=14)
plt.xlabel('City')
plt.ylabel('Number of Visits')

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.ylim(0, top_cities.values.max() * 1.1)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(
    rs['lon'],
    rs['lat'],
    s=10,
    alpha=0.5
)

plt.title('GPS Locations (Matplotlib)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```

```python
# get the count of records by date
countdata = df.groupby(df.index.date).size()
countdata.head()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    countdata.index,   # 日期
    countdata.values,  # 每天的记录数
    alpha=0.7
)

plt.title('Daily GPS Record Count (Matplotlib)')
plt.xlabel('Date')
plt.ylabel('Number of Records')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

# 确保 index 是 datetime 并排序
countdata.index = pd.to_datetime(countdata.index)
countdata = countdata.sort_index()

plt.figure(figsize=(12, 6))

plt.plot(
    countdata.index,
    countdata.values,
    linewidth=2.5,          
    color='tab:blue',      
    marker='o',             
    markersize=4,
    alpha=0.7,
    label='Daily Count'
)


max_date = countdata.idxmax()
max_value = countdata.max()

plt.annotate(
    'Peak Activity',
    xy=(max_date, max_value),          
    xytext=(max_date, max_value + 10), 
    arrowprops=dict(arrowstyle='->'),
    fontsize=10
)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45)

# 7-day rolling average
smooth = countdata.rolling(window=7).mean()

plt.plot(
    smooth.index,
    smooth.values,
    linewidth=3,
    color='red',
    alpha=0.8,
    label='7-day Rolling Average'
)

plt.title('Daily GPS Record Count with Smoothing')
plt.xlabel('Date')
plt.ylabel('Number of Records')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
